```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
path = pd.read_csv(r"C:\Users\Aaron\Downloads\archive1\videos-stats.csv")
```

```python
df1 = pd.DataFrame(path)
df1
```

```python
df1.shape
```

```python
df1.describe()
```

```python
df3 = df1.groupby(['Keyword'], as_index=False).agg({'Views':'median'})
df3 = df3.sort_values(by = 'Views',ascending = False)
df3
```

```python
sns.set(rc={"figure.figsize":(50, 4)})
ax = sns.barplot(x="Keyword", y="Views", data=df3)
```

```python
df4 = df1.groupby(['Keyword'], as_index=False).agg({'Likes':'median'})
df4 = df4.sort_values(by = 'Likes',ascending = False)
df4
```

```python
ax1 = sns.barplot(x="Keyword", y="Likes", data=df4)
```

```python
df5 = df1.groupby(['Keyword'], as_index=False).agg({'Comments':'median'})
df5 = df5.sort_values(by = 'Comments',ascending = False)
df5
```

```python
ax3 = sns.barplot(x="Keyword", y="Comments", data=df5)
```
